```python
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.uniform(-10,10,100000) #gives uniform data values from -10 to 10
plt.hist(incomes,50)
plt.show()
```

```python
#normal distribution
from scipy.stats import norm 


x = np.arange(-3,3,0.01)
plt.plot(x, norm.pdf(x))
plt.show()
```

```python
#generating random numbers with desired mean and desired standard deviation

y = np.random.normal(5,2,10000)
plt.hist(y,50)
plt.show()
```

```python
#exponential PDF "power law"

from scipy.stats import expon

x = np.arange(0,10,0.01)
plt.plot(x,expon.pdf(x))
plt.show()
```

```python
#binomial probability mass function
#mass fuuction deals with discrete data rather than continous data like PDF
from scipy.stats import binom

n,p=10, 0.5 #shape parameters
x = np.arange(0,10,0.001)
plt.plot(x,binom.pmf(x,n,p))
plt.show()
```

```python
#poisson distribution, with this we can find the odds of happening an event when we already have some prior
#knowledge of some event
#example: a site has 500 visits per day then what is the prob of getting 550 visits per day?
#here we can see that it amounts to about 0.0025

from scipy.stats import poisson

mu = 500

x = np.arange(400,600,0.5)
plt.plot(x,poisson.pmf(x,mu))
plt.show()
```
