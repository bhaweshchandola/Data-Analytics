```python
import numpy as np
import matplotlib.pyplot as plt

def de_mean(x):  #function treturn a new list which is the deviation from the mean
    xmean = np.mean(x)
    return [xi -xmean for xi in x]

def covariance(x, y): #covariance calculates the new list through mean deviation and divide by (n-1) as it is for sample
    n = len(x)
    return np.dot(de_mean(x), de_mean(y))/(n-1)

#in this example we are basically looking at how much correlation is there between a page speed and a user actually
#speding time in that web page

pagespeed = np.random.normal(3, 1, 1000)
purchaseamout = np.random.normal(50, 10, 1000)

plt.scatter(pagespeed, purchaseamout)
plt.show()

covariance(pagespeed, purchaseamout)
```

```python
purchaseamount = np.random.normal(50, 10, 1000) / pagespeed

plt.scatter(pagespeed, purchaseamount)
plt.show()

covariance(pagespeed, purchaseamount)
```

```python
'''the value of covariance is hard to interpret. so, we use correlation which makes things easy for us
and scales the value to between -1 and 1 i.e. -1 for perfect inverse relation and 1 for perfect positive correlation
. It is computed by diving the covariance by stander deviation of both the variables'''

def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x, y) / stddevx / stddevy

correlation(pagespeed, purchaseamount)
```

```python
#numpy inbuilt feature can calculate the same and returns an array with all the possible combination

np.corrcoef(pagespeed, purchaseamount)
```
