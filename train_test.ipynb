```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(2)


pagespeeds = np.random.normal(3, 1, 100)
purchaseamount = np.random.normal(50, 30, 100) / pagespeeds

plt.scatter(pagespeeds, purchaseamount, color="r")
plt.show()
```

```python
trainx = pagespeeds[:80]
testx = pagespeeds[80:]

trainy = purchaseamount[:80]
testy = purchaseamount[80:]

plt.scatter(trainx, trainy, color="r")

plt.show()
```

```python
plt.scatter(testx, testy)
plt.show()
```

```python
x = np.array(trainx)
y = np.array(trainy)

p4 = np.poly1d(np.polyfit(x, y, 8)) #here, by using 8th degree polynomial we can see that it is overfitting the data
```

```python
xp = np.linspace(0, 7, 100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0,200])
plt.scatter(x, y)
plt.plot(xp, p4(xp), c="r")
plt.show()
```

```python
x = np.array(testx)
y = np.array(testy)

axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0,200])
plt.scatter(x, y)
plt.plot(xp, p4(xp), c="r")
plt.show()
```

```python
#here we can see that r2 for test is low and for training it is much better as we did actually train it for training
from sklearn.metrics import r2_score

r2 = r2_score(testy , p4(testx))
r2
```

```python
r2 = r2_score(trainy , p4(trainx))
r2
```
